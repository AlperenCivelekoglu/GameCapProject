```python
import pandas as pd
import numpy as np

import json
import requests
from web3 import Web3
```

```python
# JSON-RPC endpoint
endpoint = 'https://eth.llamarpc.com'

# Define the JSON-RPC payload for fetching block data
payload = {
    "jsonrpc": "2.0",
    "method": "eth_getBlockByNumber",
    "params": ["0x"+format(block_number, 'x'), True],
    "id": 1
}

# Fetch block data between block numbers 17,000,000 and 17,005,000
start_block = 17000000
end_block = 17005000

blocks_data = []

for block_number in range(start_block, end_block + 1):
    response = requests.post(endpoint, json=payload)
    
    if response.status_code == 200:
        block_data = response.json().get('result')
        blocks_data.append(block_data)
        print(f"Fetched block number {block_number}")
    else:
        print(f"Failed to fetch block number {block_number}")

# Save the fetched block data in a JSON file
with open('block_data.json', 'w') as file:
    json.dump(blocks_data, file, indent=4)

print("Block data saved successfully.")
```

```python
# Read the JSON file
with open('block_data.json', 'r') as file:
    block_data = json.load(file)
```
